```python
def bubble_sort_stats(weights):
    """Return (sorted_list, swaps) where sorted_list is the bubble-sorted version of weights and swaps is the total number of swaps performed using bubble sort with early exit."""
    
    # Requirement: Work on a copy of the input list
    sorted_list = list(weights)
    n = len(sorted_list)
    
    # Requirement: Initialize swaps counter before the loops
    swaps = 0
    
    for a in range(n):
        # Requirement: Set swapped = False at start of outer pass
        swapped = False
        
        for b in range(0, n - a - 1):
            if sorted_list[b] > sorted_list[b + 1]:
                # Swap elements
                sorted_list[b], sorted_list[b + 1] = sorted_list[b + 1], sorted_list[b]
                
                # Requirement: Update flag and increment counter
                swapped = True
                swaps += 1
        
        # Requirement: Early exit if no swaps occurred in this pass
        
            
    return (sorted_list, swaps)
```

```python
def ascii_scan(s):
    result = ""
    for i in range(len(s)):
        result+= f"{s[i]} -> {ord(s[i])}"
        if i < len(s) - 1:
            result += "\n"
    return result
```

```python
def dec_to_bin8(n):
    """Return the 8-bit binary string for the integer n (0 <= n <= 255)."""
    result = ""
    for power in [128, 64, 32, 16, 8, 4, 2, 1]:
        if n >= power:
            result += '1'
            n -= power
        else:
            result += '0'
    return result
```
